```python
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion_1d(T, N, delta_t, initial_value=0):
    """
    Simulate 1D Brownian motion (Wiener process).

    Parameters:
    T (float): Total time of the simulation.
    N (int): Number of steps in the simulation.
    delta_t (float): Time increment (T/N).
    initial_value (float): The starting value of the Brownian motion.

    Returns:
    np.ndarray: Array of Brownian motion values at each time step.
    """
    dt = T / N
    increments = np.sqrt(delta_t) * np.random.normal(0, 1, N)
    W = np.concatenate(([initial_value], np.cumsum(increments)))
    return W

# Testable example: 1D Brownian motion
T = 1.0  # Total time
N = 1000  # Number of steps
delta_t = T / N  # Time increment

bm_1d = brownian_motion_1d(T, N, delta_t)

# Plotting the 1D Brownian motion
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, T, N+1), bm_1d)
plt.title('1D Brownian Motion')
plt.xlabel('Time')
plt.ylabel('W(t)')
plt.grid(True)
plt.show()
```

```python
def brownian_motion_2d(T, N, delta_t, initial_value=np.array([0, 0])):
    """
    Simulate 2D Brownian motion.

    Parameters:
    T (float): Total time of the simulation.
    N (int): Number of steps in the simulation.
    delta_t (float): Time increment (T/N).
    initial_value (np.ndarray): The starting value (position) of the Brownian motion in 2D.

    Returns:
    np.ndarray: Array of Brownian motion positions at each time step.
    """
    dt = T / N
    increments_x = np.sqrt(delta_t) * np.random.normal(0, 1, N)
    increments_y = np.sqrt(delta_t) * np.random.normal(0, 1, N)

    W_x = np.concatenate(([initial_value[0]], np.cumsum(increments_x)))
    W_y = np.concatenate(([initial_value[1]], np.cumsum(increments_y)))

    return np.vstack((W_x, W_y)).T

# Testable example: 2D Brownian motion
bm_2d = brownian_motion_2d(T, N, delta_t)

# Plotting the 2D Brownian motion
plt.figure(figsize=(10, 6))
plt.plot(bm_2d[:, 0], bm_2d[:, 1])
plt.title('2D Brownian Motion')
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.grid(True)
plt.show()
```
